Add a column to the table 

```python
# First step to make 1D array 
seats_column = pd.Series([5,6,7,6,8])  # by using pandas we add as many value in one coloumn

# one more way to add the column in the table by normal list 
Total_Seats = [2,6,4,3,4,3,8,2,7,2]  # here we have to give all value in the total row 
df["Tl_Seats"] = Total_Seats

# one more way to add the column in the table by normal list 
fuel_consume = [2,6.2,4,3,4,3.6,8,2,7.9,2]  # here we have to give all value in the total row 
df["fuel_per_km"] = fuel_consume

# Add new column named Seat
df["Seats"] = seats_column 
df
```

```python
# fill the missing value with the "5" in the seats column
df["Seats"].fillna(5,inplace = True)
df
```
